# Causal programming demo

## Pull in our dependencies

```clojure
(require '[acausal.core :refer [defmodel model->svg]])
```

## Define a model

$$w = f_w(\epsilon_w)$$
$$x = f_x(\epsilon_x)$$
$$y = f_y(z, \epsilon_y)$$
$$z = f_z(w, x, \epsilon_z)$$

$$\epsilon_w \not тлл \epsilon_z$$
$$\epsilon_x \not тлл \epsilon_y$$

```clojure
(defmodel m
  "An example structural causal model"
  {:w []
   :x []
   :y [:z]
   :z [:w :x]}
    
  #{:w :z}
  #{:x :y})
```

```clojure
(-> m :children :x)
```

```clojure
(model->svg m)
```
